```python
df.value_counts("Tumour_Stage")
```

```python
# Make a historgram of counts of people with different Tumor Stages

x1=df['Tumour_Stage']
plt.figure(figsize=(8,6))
plt.hist([x1],label=['Tumour Stage'],bins=5,alpha=0.5)
plt.legend(borderpad=0.5,fontsize=20)
plt.show()
```

```python
#Rename the Patient_Status as Target and Give labels Alive = 1 and Dead = 0

df.rename(columns={"Patient_Status": "Target"},inplace=True)
df.replace(to_replace="Alive",value="1",inplace=True)
df.replace(to_replace="Dead",value="0",inplace=True)
df.head()
```

```python
#Bin the age column
# Create a new column 'age_bins' with the Age values binned as [0, 20, 40, 60, 80, 100]
#https://pandas.pydata.org/docs/reference/api/pandas.cut.html

bins=[0, 20, 40, 60, 80, 100]

df['age_bins']=pd.cut(df['Age'],bins)
df.head()
```

```python
#plot the count of people in each age bin using a histogram

#df_r=df.value_counts("age_bins")
#df_r.hist(label=['No. of people'])

x1=df['age_bins'].value_counts()
plt.hist([x1], bins, edgecolor='black')
plt.xlabel=('Age bins')
plt.ylabel('No. of people')
plt.title('Histogram with age bins')
plt.show()
```

```python
# Make a chart showcasing number of male and female patients in the given data, explain why the data has more values of one gender over the other

new_df=df['Gender'].value_counts().rename_axis('gender').reset_index(name='count')
print(new_df)

new_df.plot.bar(x='gender', y='count',rot=0)
```
